```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
import plotly.io as pio
import plotly.colors as colors
```

```python
df = pd.read_csv(r"Sample - Superstore.csv", encoding='cp1252')
```

```python
df.head(3)
```

```python
df.info()
```

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
df["Ship Date"]=pd.to_datetime(df["Ship Date"])
df["Order Month"]=df["Order Date"].dt.month
df["Order Day of Week"]=df["Order Date"].dt.dayofweek
df["Ship Month"]=df["Ship Date"].dt.month
df["Order Year"]=df["Order Date"].dt.year
df["Ship Year"]=df["Ship Date"].dt.year
```

```python
df.columns
```

#Monthly sales Analysis

```python
sales_by_month = df.groupby("Order Month")["Sales"].sum().reset_index()
fig=px.line(sales_by_month,x="Order Month",y="Sales",title="Monthly Sales Analysis")
fig.show()
```

##Sales by Category

```python
sales_by_category = df.groupby("Category")["Sales"].sum().reset_index()
sales_by_category
```

```python
fig=px.pie(sales_by_category,names="Category",values="Sales",title="Sales by Category")
fig.show()
```

#**Sales** analysis by subcategory

```python
sales_by_subcategory=df.groupby("Sub-Category")["Sales"].sum().reset_index()
sales_by_subcategory
```

```python
fig=px.bar(sales_by_subcategory,x="Sub-Category",y="Sales",title="Sales by Sub-Category")
fig.show()
```

#**Profit** analysis by Month



```python
df["Order Month"]=df["Order Month"].map({1:"Jan",2:"Feb",3:"Mar",4:"Apr",5:"May",6:"Jun",7:"Jul",8:"Aug",9:"Sep",10:"Oct",11:"Nov",12:"Dec"})
```

```python
Sales_analysis_by_profit = df.groupby("Order Month")["Profit"].sum().reset_index()
Sales_analysis_by_profit
```

```python
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
fig = px.bar(
    Sales_analysis_by_profit,
    x="Order Month",
    y="Profit",
    title="Profit Analysis by Month",
    category_orders={"Order Month": month_order}
)
fig.show()
```

#**Profit** analysis by Category



```python
profit_analysis_by_category = df.groupby("Category")["Profit"].sum().reset_index()
profit_analysis_by_category
```

```python
fig=px.pie(profit_analysis_by_category,names="Category",values="Profit",title="Profit Analysis by Category")
fig.show()

```
